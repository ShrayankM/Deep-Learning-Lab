# McCulloch-Pitts Model 

```python
# Converting both data into numpy array
binary_train_x = binary_train_x.values
binary_test_x = binary_test_x.values
```

```python
class MPNeuron:
  
  def __init__(self):
    self.b = None
    
  def model(self, x):
    return(sum(x) >= self.b)
  
  def predict(self, X):
    Y = []
    for x in X:
      result = self.model(x)
      Y.append(result)
    return np.array(Y)
  
  def fit(self, X, Y):
    accuracy = {}
    
    for b in range(X.shape[1] + 1):
      self.b = b
      Y_pred = self.predict(X)
      accuracy[b] = accuracy_score(Y_pred, Y)
      
    best_b = max(accuracy, key = accuracy.get)
    self.b = best_b
    
    print('Optimal value of b is : ', best_b)
    print('Highest accuracy of y_train data : ', accuracy[best_b])
```

# Perceptron model 

```python
class Perceptron:
    # weight and bias is set to none in __init__
    def __init__(self, learning_rate=0.01, iterations = 1000):
        self.bias = None
        self.weights = None
        self.activation_func = self.unit_step_func
        self.lr = learning_rate
        self.iters = iterations
        
    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = self.activation_func(linear_output)
        return y_predicted
        
    def unit_step_func(self, x):
        return np.where(x>=0,1,0)
    
    def fit(self, X , y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        # y is either 0 or 1
        y_ = np.array([1 if i>0 else 0 for i in y])
        
        for itr in range(self.iters):
            for indx, x_i in enumerate(X):
                # linear_output = x*weight + bias
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = self.activation_func(linear_output)
                # update is learning_rate *y[index] - y_predicted
                update = self.lr * (y_[indx] - y_predicted)
                # weight is updated as x * update
                self.weights += update * x_i
                # bias is updated as bias + update
                self.bias += update
```

```python
def accuracy(y_testing, y_pred):
    accuracy = np.sum(y_testing == y_pred) / len(y_testing)
    print(accuracy)
    return accuracy
```

```python
# convert x_train, x_test into arrays
x_train = x_train.values
x_test = x_test.values
```

```python
p = Perceptron(learning_rate = 0.02, iterations=1000)
p.fit(x_train,y_train)
```

```python
prediction = p.predict(x_test)
print("Perceptron model accuracy is : ",accuracy(y_test, prediction))
```
